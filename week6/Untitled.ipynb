```python
os.path.isfile("Happiness.csv")
filepath = "Happiness.csv"
    
happy_data = pd.read_csv(filepath)
print "done"
```

```python
hdgdp = happy_data.sort_values(by=["Economy (GDP per Capita)"],ascending=[False])
hdgdp[:20]
```

```python
pie2 = hdgdp[:20]["Region"].value_counts()
pie2.plot(kind="pie")
```

```python
hdgdp[:20].plot.bar(x="Country", y="Economy (GDP per Capita)")
```

```python
hdgdp[:20].plot.bar(x="Country", y="Happiness Score", color="red")
hdgdp[:20].plot.bar(x="Country", y="Happiness Rank", color="yellow")
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)
ax2 = ax.twinx()
hd = hdgdp[:20]

hd.plot.bar(x="Country", y="Economy (GDP per Capita)", color='blue', ax=ax, position=1)
hd.plot.bar(x="Country", y="Happiness Score", color='red', ax=ax2, position=0)

ax.set_ylabel('econ')
ax2.set_ylabel('happy')
ax.set_xlabel('Country')

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)
ax2 = ax.twinx()
hd = hdgdp[:20]

hd.plot.bar(x="Country", y="Economy (GDP per Capita)", color='blue', ax=ax, position=1)
hd.plot.bar(x="Country", y="Happiness Rank", color='yellow', ax=ax2, position=0)

ax.set_ylabel('econ')
ax2.set_ylabel('rank')
ax.set_xlabel('Country')

plt.show()
```
